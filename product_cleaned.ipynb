```python
import pandas as pd
from sklearn import linear_model, model_selection
import numpy as np
# test
products = '/Users/paddy/Desktop/facebook_products.csv'

df =  pd.read_csv(products)
df.head()
```

```python
df.dtypes
```

```python
df[['id','product_name','category','product_description','location','url','create_time']] = df[['id','product_name','category','product_description','location','url','create_time']].astype(str)
df.dtypes
```

```python
df = df.drop(df.columns[0], axis=1)
df['product_name'] = df['product_name'].str.encode('ascii', 'ignore').str.decode('ascii')
df['product_description'] = df['product_description'].str.encode('ascii', 'ignore').str.decode('ascii')
df.head()
```

```python
df = df.dropna(axis=0)
df.info
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace('£','')
df['price'] = df['price'].str.replace(',','').astype(float)
df.head()
```

```python
df.dtypes
```

```python
# df['product_name'] = df['product_name'].astype('str')
# df['product_name'] = df['product_name'].str.replace('|','')

df['product_name'] = df['product_name'].str.split('|').str[0]
# df['product_name'] = df['product_name']
df.head()
```

```python
df['subcategory'] = df['category'].str.split('/').str[1]
df['category'] = df['category'].str.split('/').str[0]
df.head()
```

```python
df['county'] = df['location'].str.split(',').str[1]
df['location'] = df['location'].str.split(',').str[0]
df.head()
# fin
```
